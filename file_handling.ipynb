# File Handling

File Handling in Python lets us read, write and manipulate files stored in the computer memory.

Reading an already existing file.

`f=open("sample.txt","r")`

this creates a file object 'f'. Only existing files can be read through read mode 'r'. Trying to read a non existing file throws an error

`f=open("sample.txt","w")`

if a non existing file is being opened using write mode 'w', a new file with the given name is created.

https://www.geeksforgeeks.org/python/file-handling-python/

Common file modes:
- 'r' → read
- 'w' → write (overwrites)
- 'a' → append
- 'r+' → read and write

This reference contains all the file modes: https://www.geeksforgeeks.org/python/file-mode-in-python/



```python
f=open("sample.txt",'r')
```

```python
#read method is used to read the contents of a file
#readline can be used to read one single line at a time
#readlines can be used to read all the content of the file, this method returns a list contain each line as an element
print(f.read())
```

## Using `with open()`

in order to skip closing the file everytime we can make use of the `with open()` function which closes the file automatically after execution

```python
with open("sample.txt",'w') as file:
    file.write("Updated content using write function")
print("File Updated")
```

```python
with open("sample.txt",'a') as file:
    file.write("\n Sample file for demonstrating File Handling")
print("File Appended")
```

We can make use of the OS module to check the existence of a file or to delete a file

```python
import os
if os.path.exists("sample.txt"):
    print("File Exists")
else:
    print("File Doesn't Exist")
```

`os.remove("sample.txt")` will delete sample.txt
