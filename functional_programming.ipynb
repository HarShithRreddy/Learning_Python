# Functional Programming

## List Comprehension

List comprehensions provide a concise and readable way to create new collections by applying an expression to each item in an iterable.

```python
#Normal way
table=[]
for i in range(1,11):
    table.append(2*i)
print(table)
```

```python
#Using List Comprehension
t=[2*i for i in range(1,11)]
print(t)
```

We can do the same with dictionaries,sets.

```python
t_dict={x: 2*x for x in range(1,11)}
print(t_dict)
```

```python
#We can also add conditions in comprehensions
t_2=[2*i for i in range(1,11) if i%2==0]
print(t_2)
```

## Lambda Functions

It is an anonymous, single-expression function defined using the lambda keyword. It is used for short throw-away functions where a full function definition is unnecessary

```python
square=lambda x:x*x
print(square(7))
```

```python
divisible=lambda x,y:x%y==0
print(divisible(6,3))
```

## Other Functional aspects:

- map():Map is used to apply a function to all elements of a data structure.
- filter():Filter is used to check each element of a data structure with a condition

These functions are lazy functions, they return an object instead of a proper data structure. These objects have to be explicitly converted to be printed

```python
#map function
numbers=[1,2,3,4,5]
squared=list(map(lambda x:x**2,numbers))
print(squared)
```

```python
#filter function
numbers=[1,2,3,4,5]
print(list(filter(lambda x:x%2==0,numbers)))
```
