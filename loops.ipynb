# Loops in Python

## for loop
This type of loop is used to iterate over a data structure like lists,tuples and dictionaries.

for loops can also be used to print numbers in a range excluding the ending number.

Syntax:

`for i in range(start,end,skip)`

```python
o=[1,3,5,7,9]
for i in o:
    print(i) #i is a looping variable used to iterate over the list o
```

```python
for j in range(0,11,2):
    print(j) #prints all even numbers from 0 to 10
```

## While Loop
while loops are used to loop until a given specific condition is true.

Syntax:

`while(condition)`

Indentation is crucial in Python and defines the scope of the loop body.

```python
i=0
while(i<=15):#prints numbers divisible by 3 until 15
    if(i%3==0):
        print(i)
    i+=1
```

## Loop Control Statements
- `pass`: this is used as a placeholder inside a function or a loop. pass does nothing when encountered and the compiler just moves on to the next line of code.
- `continue`: continue is used to skip the current iteration of the loop and go to the next iteration
- `break`: break is used to exit the function or loop completely

```python
def empty():
    pass #this function can be filled later based on the next part/output of the code.
```

```python
p=[1,2,3,4,5]
for k in p:
    if(k==3):
        continue #skips 3
    print(k)
```

```python
for m in range(1,11,3):
    if (m==7):# breaks out of the loop when m becomes 7
        break
    print(m)
```
