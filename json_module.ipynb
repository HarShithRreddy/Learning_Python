# JSON 

JavaScript Object Notation(JSON) is a lightweight data format used to store and exchange data between servers and applications.

Python has a inbuilt json module to work on json objects.

Importing json module:

`import json`

JSON supports only basic data types (string, number, boolean, null, arrays, objects).

```python
import json
obj={"name": "Harsh" ,"gender": "Male" } #creating a json object ultimately a dictionary in python
print(obj)
```

```python
#Using the dumps method which converts the python object into a json string
str=json.dumps(obj)
print(str)
type(str) #It will show up as a string 
```

```python
#loads method converts a json string to a python object
d=json.loads(str)
print(type(d))
print(d)
```

```python
#we can use the dump and load methdods to work with reading and writing json modules into a file
with open("data.json",'w') as file:
    json.dump(obj,file)
print("Json data dumped succesfully")
```

```python
with open("data.json",'r') as file:
    s=json.load(file)
    print(s)
```

```python
#we can also use dumps for pretty printing a json object
pretty = json.dumps(obj, indent=4)
print(pretty)
```
