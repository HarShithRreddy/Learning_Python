# Data Structures in Python
## Lists
lists in Python are a mutable pair of collection of different type of elements.

`my_numbers=[1,'Harshith',3.14]`

we depict a empty list with square brackets

`empty_numbers=[]`

For a complete reference of list methods, external documentation can be consulted. https://www.geeksforgeeks.org/python/numbers-methods-python/

```python
#Accesing list elements
numbers=[1,3,5,7,9]
print(numbers[0])
print(numbers[-1])
```

```python
#lists are mutable can change the values using indexes
print("The old numbers is",numbers)
numbers[0]=2
print("The new numbers is",numbers)
```

```python
#Appending a element into the list
numbers.append(11)
print(numbers)
```

```python
#Concatenating another list 
numbers2=[4,6]
numbers.extend(numbers2)
print(numbers)
```

```python
#Inserting an element at a specific index
numbers.insert(2,8)
print(numbers) #insert method shifts the remaining elements after the index to the right
```

```python
#Removing an element
#There are two methods remove() and pop()
numbers.remove(9) #remove method takes the element to be removed and doesnt return the element
print(numbers)
removed=numbers.pop(2) #pop method takes the index to be removed and left shifts the elements after it in the ;ist. It also returns the element removed. If no index is provided it removes the element at last index
print("The removed element is",removed)
print(numbers)
```

```python
#index method returns the index of the first occurence of the given element
print(numbers.index(5)) #returns 2
print(numbers.index(6)) #returns 6
```

```python
#count method returns the number of times a given element occurs in the list
numbers.append(2)
print(numbers.count(2)) #returns 2
print(numbers.count(3)) #returns 1
```

```python
#sort method sorts the given elements in ascending order(alphabetical incase of strings)
numbers3= [2,85,93,444,3,4]
numbers3.sort()
print(numbers3)
#sorted method also sorts the elements of a numbers but returns a new numbers instead of sorting the numbers itself
numbers4=[2,85,93,444,3,4]
sorted_numbers=sorted(numbers3)
print(sorted_numbers)
```

```python
#a list can also be sliced like a string
print(numbers4[0:3])
```

Other methods of numbers are as follows:
- `numbers.reverse()`: reverses a given numbers
- `numbers.copy()`: creates a copy of the numbers
- `numbers.clear()`: clears the entire numbers

```python
# a string with multiple elements with a delimiter can be converted to a list using the split method
s="jack-jill-went-up-the-hill"
str=s.split("-")
print(str)
```

```python
#we can also have nested lists (list inside a list)
nested_list=[1,2,3,[4,5]]
#the elements inside a nested list can be accessed using double indexing
print(nested_list[3][1])
```

## Tuples

A tuple in python is an ordered, immutable pairof collection. They are slightly faster than lists.

Creating a tuple:

`my_tuple=(5,8,3,2)`

Creating an empty tuple:

`empty_tuple=()`

For reference of all the tuple methods: https://www.geeksforgeeks.org/python/python-tuples/

```python
#elements of the tuple can be accessed like in the lists
t=(0,9,"Hello","World")
print(t[2])
```

This will raise an error because tuples are immutable

 `t[0] = 10`

```python
#len method to find the length of the tuple
odd=(1,3,5,7,9)
print(len(odd))
```

Packing and Unpacking can be done using tuples

```python
#Packing
Customer=("Alice",25,"New York City")
#Unpacking
name,age,place=Customer
print(f"{name} is {age} years old and lives in {place}")
```

All of the features and methods like slicing, index, count etc.. work in the same way as in lists but only with the conditions that tuples cannot be changed.

Tuples are very useful in cases where u need a collection that doesn't change in the course of execution of the program.

# Sets

A set is an undordered, mutable and an unique set of collection in python. It does not allow duplicate set of values.

Creating a set:

`my_set={2,4,6,8}`

Creating an empty set:

`empty_set=set()`

using `{}` to create an empty set isn't correct because it creates an empty dictionary which we will look into later.

For reference of all the set methods: https://www.geeksforgeeks.org/python/python-set-methods/

```python
#we can create a set using a list and vice versa
my_set={2,4,6,8}
my_list=list(my_set)
print(my_list)
#The output is in random order because sets are unordered and so the converted list also tends to be unordered
```

```python
#set doesn't allow duplicate values
s={1,1,2,3,4,5,5}
print(s)
```

```python
#add method adds a new element to the set
s.add(7)
print(s)
```

```python
#To add multiple elements we make use of the update method. The argument for the update method should be a list
s.update([1,9,6,"Harsh"])
print(s)
```

```python
#remove and discard methods are used to remove an element from a set. The only difference between remove and discard is that remove raises an error if the given element doesn't exist in the set
#pop method can also be used to remove a random element from the set and returns it
s.remove("Harsh")
print(s)
s.discard(3)
print(s)
```

### Real World Set operations inside python

- Union :  Returns the a new set with combined two sets. `set1.union(set2)`
- Intersection: Returns a new set with the common elements of both sets. `set1.intersection(set2)`
- Difference: Returns a new set with elements in set1 but not in set2. `set1.difference(set2)`
- Symmetric_Difference: Returns a set with non common elements in both sets. `set1.symmetric_difference(set2)`
- Subset: Returns `True` if set1 is subset of set2 or returns `False`. `set1.issubset(set2)`
- Superset: Returns `True` if set1 is superset of set2 or returns `False`. `set1.issuperset(set2)`
