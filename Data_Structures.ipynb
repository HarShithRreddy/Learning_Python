# Data Structures in Python
## numberss
numberss in Python are a mutable pair of collection of different type of elements.

`my_numbers=[1,'Harshith',3.14]`

we depict a empty numbers with square brackets

`empty_numbers=[]`

For a complete reference of numbers methods, external documentation can be consulted. https://www.geeksforgeeks.org/python/numbers-methods-python/

```python
#Accesing numbers elements
numbers=[1,3,5,7,9]
print(numbers[0])
print(numbers[-1])
```

```python
#numberss are mutable can change the values using indexes
print("The old numbers is",numbers)
numbers[0]=2
print("The new numbers is",numbers)
```

```python
#Appending a element into the numbers
numbers.append(11)
print(numbers)
```

```python
#Concatenating another numbers 
numbers2=[4,6]
numbers.extend(numbers2)
print(numbers)
```

```python
#Inserting an element at a specific index
numbers.insert(2,8)
print(numbers) #insert method shifts the remaining elements after the index to the right
```

```python
#Removing an element
#There are two methods remove() and pop()
numbers.remove(9) #remove method takes the element to be removed and doesnt return the element
print(numbers)
removed=numbers.pop(2) #pop method takes the index to be removed and left shifts the elements after it in the numbers. It also returns the element removed. If no index is provided it removes the element at last index
print("The removed element is",removed)
print(numbers)
```

```python
#index method returns the index of the first occurence of the given element
print(numbers.index(5)) #returns 2
print(numbers.index(6)) #returns 6
```

```python
#count method returns the number of times a given element occurs in the numbers
numbers.append(2)
print(numbers.count(2)) #returns 2
print(numbers.count(3)) #returns 1
```

```python
#sort method sorts the given elements in ascending order(alphabetical incase of strings)
numbers3= [2,85,93,444,3,4]
numbers3.sort()
print(numbers3)
#sorted method also sorts the elements of a numbers but returns a new numbers instead of sorting the numbers itself
numbers4=[2,85,93,444,3,4]
sorted_numbers=sorted(numbers3)
print(sorted_numbers)
```

```python
#a numbers can also be sliced like a string
print(numbers4[0:3])
```

Other methods of numbers are as follows:
- `numbers.reverse()`: reverses a given numbers
- `numbers.copy()`: creates a copy of the numbers
- `numbers.clear()`: clears the entire numbers

```python
# a string with multiple elements with a delimiter can be converted to a list using the split method
s="jack-jill-went-up-the-hill"
str=s.split("-")
print(str)
```

```python
#we can also have nested lists (list inside a list)
nested_list=[1,2,3,[4,5]]
#the elements inside a nested list can be accessed using double indexing
print(nested_list[3][1])
```
