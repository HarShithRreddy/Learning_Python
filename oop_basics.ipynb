# Object Oriented Programming (OOP) in Python

Object Oriented Programming (OOP) is a programming paradigm based on the concept of objects.

An object:
- Represents a real-world entity
- Contains data (attributes)
- Contains behavior (methods)

A class is a blueprint for creating objects.

```python
#Creating a class for the students of 12th grade
class Students:
    grade=12 #This is a class attribute, since it is same for all the objects
    def printDetails(self,name,roll_no,total_marks): #self is a variable that represents the object itself
        print(f"{name} bearing roll no {roll_no} has obtained {total_marks} in the final examination")# these are instance variables which vary depending on the object
```

```python
Harsh=Students() #creating different objects which follow the template of the class
Harsh.printDetails("Harsh",12,567)# all of the functions defined in the class can be used by all its objects
John=Students()
John.printDetails("John",23,551)
```

```python
#We can use a constructor inside a class to make things easier to pass the arguments
class Students2:
    def __init__(self,name,roll_no,total_marks):#constructor function
        self.name=name
        self.roll_no=roll_no
        self.total_marks=total_marks
    def printDetails(self):
        print(f"{self.name} bearing roll no {self.roll_no} has obtained {self.total_marks} in the final examination")
```

```python
Harsh=Students2("Harsh",12,567)#arguments can be passed during object creation
Harsh.printDetails()
```

```python
class Students3:
    grade=12
    def __init__(self,name,roll_no,total_marks):#constructor function
        self.name=name
        self.roll_no=roll_no
        self.total_marks=total_marks
    def printDetails(self):
        print(f"{self.name} bearing roll no {self.roll_no} has obtained {self.total_marks} in the final examination")
#every function of a class must bear self, but if we want omit it we can use the static method
    @staticmethod
    def printGrade():
        print("This is Grade 12")
#we can use class method to access class variables directly
    @classmethod
    def printGrade2(cls):
        print(f"This is grade{cls.grade}")
```

```python
purav=Students3("Purav",31,591)
purav.printDetails()
purav.printGrade()#would throw an error if staticmethod wasn't used
purav.printGrade2()
```
