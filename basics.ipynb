# Operators in Python

There are six types of operators in python 
- Arithmetic(+,-,*,/,,//,%,**) : used for basic arithmetic operations
- Relational(==,!=,>,<,>=,<=,) : used to compare variables
- Logical(and,or,not): used to combine conditional statements 
- Bitwise(&,|): used to perform bit operations
- Assignment(shorthand) (=,+=,-=,*=,/=,//=,%=,**=) : used to assign variables 
- Membership (in, not in , is , is not): checks for presence of an object or entity in a list

```python
a = 10
b = 3
print(a + b, a * b, a ** b)
```

```python
print(a > b)
print(a == b)
```

```python
print(a > 5 and b < 5)
```

```python
a+=2
print(a)
```

```python
print(3 in [1, 2, 3])
```

# Operator Precedence in Python

Python follows the PEMDAS order.
1. Parenthesis ()
2. Exponents** evaluated right to left
3. Multiplication *, Division /, Floor Division // , Modulus %, evaluated left to right
4. Addition +, Subtraction -, evaluated left to right

```python
result = 10 + 2 * 3
print(result) # multiplication has higher precedence so (2*3) happens first and then gets added to 10
```

# Conditional Statements

Python uses `if`, `elif`, and `else` to execute code based on conditions.

- `if` is used to check the initial condition.
- `elif` (else if) is used to check additional conditions when the previous condition is false.
  There can be multiple `elif` blocks.
- `else` is used to execute code when none of the above conditions are true.

```python
x = 10

if x > 10:
    print("x is greater than 10")
elif x == 10:
    print("x is equal to 10")
else:
    print("x is less than 10")
```

# Match Case
Used instead of conditional statements in cases where a single variable must be matched to its respective case which executes a particular block of code.

```python
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:
            return "Unknown Status"

print(http_status(200))  # Output: OK
print(http_status(404))  # Output: Not Found
```

# String Formatting and f-strings

Strings are immutable, once defined they cannot be changed.

f-strings can be used inside the print function to format the output with the variable values

```python
def greet(name):
    print(f"Hello Mr./Mrs. {name}");
greet("Harshith")
print(f"The sum of 2 and 3 is {2+3}")
```

```python
#We can align the outputs using string padding
print(f"{'Python':<10}")  # Left-align
print(f"{'Python':>10}")  # Right-align
print(f"{'Python':^10}")  # Center-align
```
